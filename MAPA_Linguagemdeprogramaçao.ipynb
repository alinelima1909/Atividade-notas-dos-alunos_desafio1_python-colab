Laço While

```python
pedidos = [
    {"id": 101, "status": "pago", "valor": 120.0, "cancelado": False},
    {"id": 102, "status": "pago", "valor": 80.0,  "cancelado": True},
    {"id": 103, "status": "pendente", "valor": 50.0, "cancelado": False},
    {"id": 104, "status": "pago", "valor": 40.0,  "cancelado": False},
]
alturas = [55, 48, 50, 39, 72]  # em centímetros
limite = 100.0
altura_minima_alta = 50
```

```python
i = 0
total = 0
validos = 0

while i < len(pedidos) and pedidos[i]["status"] == "pago":
    total += pedidos[i]["valor"]
    if not pedidos[i]["cancelado"]:
        validos += 1
    i += 1
```

```python
i = 0
total = 0
validos = 0

while i < len(pedidos):
    pedido = pedidos[i]

    # Só processa pedidos pagos e não cancelados
    if pedido["status"] == "pago" and pedido["cancelado"] == False:
        total += pedido["valor"]
        validos += 1
    else:
        break  # Interrompe ao encontrar condição fora do padrão

    i += 1  # Incremento obrigatório
```

```python
print(f"Total: {total}")
print(f"Validos: {validos}")
```

Parte For

```python
contagem_altos = 0

# Conta quantas alturas são <= 50
for j in range(len(alturas)):
    if alturas[j] <= 50:
        contagem_altos += 1
```

```python
contagem_altos = 0

for altura in alturas:
    # Validação para ignorar valores inconsistentes
    if isinstance(altura, (int, float)) and altura > 0:
        if altura <= 50:
            contagem_altos += 1
```
