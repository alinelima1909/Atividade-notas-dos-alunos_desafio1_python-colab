Segurança	5	3000
Docente	16	6000
Diretoria	1	12500
Cinco pessoas trabalham como seguranças e recebem um salário de 3 mil reais. Dezesseis pessoas trabalham como docentes e recebem um salário de 6 mil reais. Por fim, temos uma pessoa que trabalha na diretoria e seu salário é de 12 mil e 500 reais.

```python
q_segurança = 5
s_segurança = 3000

q_docente = 16
s_docente = 6000

q_diretoria = 1
s_diretoria = 12500
```

```python
total_de_empregados = q_segurança + q_docente + q_diretoria
total_de_empregados
```

```python
diferença_de_salario = s_segurança - s_diretoria
diferença_de_salario
```

```python
media = (q_segurança*s_segurança + q_docente*s_docente + q_diretoria*s_diretoria) / (total_de_empregados)
media
```
